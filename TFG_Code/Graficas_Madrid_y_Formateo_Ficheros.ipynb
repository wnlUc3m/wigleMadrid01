```python
  def df_to_json(dataframe):
        data = [];
        for t in dataframe.itertuples():
            
            line = {"nombre": t.nombre_distrito, "porcentaje_movistar":t.movistar/float(t.total), 
                    "porcentaje_vodafone":t.vodafone/float(t.total), "porcentaje_orange":t.orange/float(t.total),
                   "porcentaje_jazztel":t.jazztel/float(t.total), "porcentaje_ono":t.ono/float(t.total), 
                    "porcentaje_wlan":t.wlan/float(t.total)};
            data.append(line);
            
        return data;   
    
def save_json(data):
    with open( "datos.json", 'w') as outfile:    
        json.dump(data, outfile);  
```
